```python
# people doing job in country
country_job = df["Country"].value_counts()

# f=["LanguageHaveWorkedWith"].str.contains("Python").value_counts()
# print(f)

print(country_job)
```

```python
country_use_python= group["LanguageHaveWorkedWith"].apply(lambda x:x.str.contains("Python").sum())
country_use_python
```

```python
# it return the no. of people work in a country in specific language
python_df=pd.concat([country_job,country_use_python],axis='columns',)
python_df
```

```python
python_df.rename(columns={"count":"employee count" ,"LanguageHaveWorkedWith" :"python employees"}, inplace=True)

# type(df.loc[0,"employee count"])
```

```python
# add the python employee percentage in the country
python_df["python_emp%"] = (python_df["python employees"] / python_df["employee count"])*100

python_df
```

```python
# sort according to the any particular column
python_df.sort_values(by="Country" , inplace=True)
python_df
```
