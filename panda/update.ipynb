```python
people = {
    "first": ["yash", 'riya', 'khushi'], 
    "last": ["jain", 'goyal', 'sathpute'], 
    "email": ["yashjain@gmail.com", 'riyagoyal@email.com', 'khushi@email.com']
}
```

```python

import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

Ways of the updating tabel

```python
df.columns
```

```python
# first way
# its want to put all the columns name in a array
df.columns=['first_name', 'last_name', 'email']
df
```

```python
# secound way of updating
df.columns=df.columns.str.replace("_"," ")
df
```

```python
# third way
# update in data directly
df.columns = [x.upper() for x in df.columns]
df
```

```python
# fourth way 
# not update data whthout a inplace=True
df.rename(columns={"FIRST NAME":'first' , "LAST NAME":'last'},inplace=True)
df
```

```python
# fifth way
# it want a all the columns value 
# df.loc[2] = ['John', 'Smith'] it is not valid show error
df.loc[2] = ['John', 'Smith', 'JohnSmith@email.com']
df
```

```python
# sixth way 
# pick row than the column of the row = updated value
df.loc[[2], ['last', 'EMAIL']] = ['Doe', 'JohnDoe@email.com']
df
```

```python
# seventh way 
# df.loc[2,"last"]="yash" 
# both are the work same
df.at[2,"last"]="yash"
df
```

```python
#  it give a error
# filt = (df['email'] == 'JohnDoe@email.com')
# df[filt]['last'] = 'Smith'
```

```python
df.loc[1,"EMAIL"]="JohnDoe@email.com"

filt = (df['EMAIL'] == 'JohnDoe@email.com')
df.loc[filt, 'last'] = 'Smith'
df
```

```python
df.columns=df.columns.str.lower()
df
```
