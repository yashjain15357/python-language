```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df = pd.DataFrame(people)
df
```

Filter the data 

```python
# 
filt = df["last"]=='Doe'
df[filt]
```

```python
print(df.loc[filt])
```

```python
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
df.loc[filt , ['first', 'last']]
```

```python
df.loc[~filt, 'email']
```
