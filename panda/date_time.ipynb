```python
import pandas as pd
from datetime import datetime
```

```python
d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d %I-%p')
df = pd.read_csv(r"C:\Users\YASH JAIN\Documents\New Folder\OneDrive\Desktop\CodeBox\python language\panda\data\ETH_1h.csv" , parse_dates=["Date"] , date_parser=d_parser)
df
```

```python
df.loc[0,"Date"]
```

```python
# it return the day of the date
df.loc[0,"Date"].day_name()
```

%Y-%m-%d → Year-Month-Day (2020-03-13)

%I → 12-hour format hour (08)

%p → AM/PM indicator (PM)

- → Literal hyphen between hour and AM/PM

```python
df["Date"] = pd.to_datetime(df["Date"] , format="%Y-%m-%d %I-%P")
df
```

```python
# Adding a new column 'DayofWeek' that contains the name of the day for each date
df["DayofWeek"] = df["Date"].dt.day_name()
df
```

```python
# starting/ smallest date return 
df['Date'].min()
```

```python
# last/ highest date return 
df['Date'].max()
```

```python
# difference the starting and last date 
df['Date'].max() - df['Date'].min()
```

```python
# filter the data acc. to the date
filt = (df["Date"]>="2020")
df.loc[filt]
```

```python
# filter the according to the year
filt = (df["Date"]< "2020") &(df["Date"]< "2019")
df.loc[filt]
```
