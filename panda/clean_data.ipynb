```python
import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
# Remove all rows that contain NaN values from the dataset
df.dropna()
# it needs a inplace=True for the change in a dataset
```

NOTE: it the survey data use a NA value Missing value etc so for cleaning data first replace NA or Missing etc. value with NaN

```python
df.replace("NA",np.nan,inplace=True)
df.replace("Missing",np.nan,inplace=True)
df.dropna()
```

```python
# it return a tabel in boolean for 
# it replace the NaN vale with the True and non-NaN value with False
df.isna()
```

```python
df
```

```python
# "any" (default) → Drops rows/columns if any value is NaN
# 0 (default) → Removes rows with NaN
df.dropna(axis=0, how='any',  inplace=False)
```

```python
# "all" → Drops rows/columns only if all values are NaN
# 1/"columns" → Removes columns with NaN
df.dropna(axis="columns", how='all',  inplace=False)
```

```python
# "all" → Drops rows/columns only if all values are NaN
# 1/"columns" → Removes columns with NaN

# subset:-check for NaN (instead of all)

df.dropna(how='any', subset=["email","age"], inplace=False)
```

```python
# fill a  NaN value with another value all the data type
df.fillna(-1)
```
